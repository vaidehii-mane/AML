Import libraries and load dataset

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris

data=load_iris()
X=pd.DataFrame(data.data,columns=data.feature_names)
y=pd.Series(data.target)

print(X.head())
print(y.head())
```

Filter Methods

1. Correlation Coefficient Method

```python
cor=X.corrwith(y)
print("Correlation scores: ",cor)
```

2. Variance Threshold Method

```python
from sklearn.feature_selection import VarianceThreshold

selector = VarianceThreshold(threshold=0.6)
X_variance = selector.fit_transform(X)

print("Selected features shape:", X_variance.shape)
```

3. Chi-Square Feature Selection

```python
from sklearn.feature_selection import SelectKBest, chi2

X_positive = X - X.min()

chi2_selector = SelectKBest(score_func=chi2, k=2)
X_chi2 = chi2_selector.fit_transform(X_positive, y)

print("Chi-Square selected features shape:", X_chi2.shape)
print("Chi-Square scores:", chi2_selector.scores_)
```

4. ANOVA F-Test

```python
from sklearn.feature_selection import f_classif

anova_selector = SelectKBest(score_func=f_classif, k=2)
X_anova = anova_selector.fit_transform(X, y)

print("ANOVA selected features shape:", X_anova.shape)
print("ANOVA F-scores:", anova_selector.scores_)
```

5. Information Gain (Mutual Information)

```python
from sklearn.feature_selection import mutual_info_classif

info_gain = mutual_info_classif(X, y)

for feature, score in zip(X.columns, info_gain):
    print(feature, ":", score)
```
